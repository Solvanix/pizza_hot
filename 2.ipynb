```python
from bs4 import BeautifulSoup

# تحميل الصفحة
with open("/home/ali/0_gh_repos/pizza_hot/index.html", "r", encoding="utf-8") as f:
    html = f.read()

soup = BeautifulSoup(html, "html.parser")

# قائمة الصور المحلية التي نريد استخدامها بدل صور Ecwid
replacement_images = [
    "index_files/pizza_closeup_main.png",
    "index_files/pizza_brick_oven.jpg",
    "index_files/wings_parsley_lime_plate.jpg"
]

# مؤشر لتدوير الصور المحلية
img_counter = 0

# استبدال الصور داخل <img src="...">
for img in soup.find_all("img"):
    src = img.get("src", "")
    if "cloudfront.net" in src:
        img["src"] = replacement_images[img_counter % len(replacement_images)]
        img_counter += 1

# استبدال الصور داخل style="background-image: url(...)"
for tag in soup.find_all(style=True):
    style = tag["style"]
    if "background-image" in style and "cloudfront.net" in style:
        start = style.find("url(")
        end = style.find(")", start)
        if start != -1 and end != -1:
            new_url = f"url('{replacement_images[img_counter % len(replacement_images)]}')"
            tag["style"] = style[:start] + new_url + style[end+1:]
            img_counter += 1

# حفظ التعديلات
with open("/home/ali/0_gh_repos/pizza_hot/index.html", "w", encoding="utf-8") as f:
    f.write(str(soup))

print(f"✅ تم استبدال {img_counter} صورة من Ecwid بصور محلية دون المساس بالتنسيقات.")
```

```python
from bs4 import BeautifulSoup

html_path = "/home/ali/0_gh_repos/pizza_hot/index.html"

# تحميل الصفحة
with open(html_path, "r", encoding="utf-8") as f:
    html = f.read()

soup = BeautifulSoup(html, "html.parser")

# البحث عن الهيدر المحتمل (بناءً على tag أو class أو وجود كلمات مفتاحية)
possible_headers = []

for tag in soup.find_all(True):
    # أي عنصر فيه روابط تنقل أو شعار
    text = tag.get_text(strip=True)
    if any(x in text for x in ["عن", "قائمة", "السلة", "Pizza Hot", "الصفحة الرئيسية"]):
        if tag.name in ["header", "nav", "div", "section"] and tag not in possible_headers:
            possible_headers.append(tag)

# نحاول إصلاح أول عنصر فقط لضمان الحذر
if possible_headers:
    header = possible_headers[0]

    # إذا كان الهيدر داخل a أو section غير مناسب → نخرجه
    parent = header.find_parent()
    if parent.name == "a" or parent.name == "section":
        parent.unwrap()

    # ننقله لأعلى body إذا لم يكن هناك بالفعل
    if soup.body and header not in soup.body.contents[:2]:
        header.extract()
        soup.body.insert(0, header)

    # حفظ النتيجة
    with open(html_path, "w", encoding="utf-8") as f:
        f.write(str(soup))

    print("✅ تم استعادة الهيدر إلى أعلى الصفحة بنجاح.")
else:
    print("⚠️ لم يتم العثور على عنصر يُشبه الهيدر. يرجى التحقق يدويًا أو مشاركتي بنية الصفحة.")
```

```python
from bs4 import BeautifulSoup
from datetime import datetime

html_path = "/home/ali/0_gh_repos/pizza_hot/index.html"

# قراءة الصفحة
with open(html_path, "r", encoding="utf-8") as f:
    html = f.read()

soup = BeautifulSoup(html, "html.parser")

# 🧹 1. حذف كل العناصر النصية المتعلقة بالتذييل باستثناء الحقوق
keywords_to_remove = [
    "مفتوح يوميًا", "نسيم", "Pizza Hot Park", 
    "اطلب الآن", "نخبز الشغف", "تحسين تجربتكم", "طابع فلسطيني",
    "بلعين", "رام الله", "زورونا", "التوصيل", "جميع الحقوق"
]

for tag in soup.find_all(string=True):
    if any(kw in tag for kw in keywords_to_remove):
        parent = tag.find_parent()
        if parent and "©" not in tag:
            try:
                parent.decompose()
            except:
                continue

# 🧱 2. حذف أي تذييل موجود بالكامل
for footer_tag in soup.find_all("footer"):
    try:
        footer_tag.decompose()
    except:
        continue

# 🆕 3. إنشاء تذييل جديد يحتوي فقط على سطر الحقوق
footer = soup.new_tag("footer", attrs={
    "style": "background:#fff; color:#222; padding:20px 0; text-align:center; font-size:14px; font-family:Tajawal, Arial;"
})

rights_text = f"© جميع الحقوق محفوظة Pizza Hot {datetime.now().year}"
p = soup.new_tag("p")
p.string = rights_text
footer.append(p)

# 📌 إلحاق التذييل الجديد في نهاية body
body = soup.body or soup.find("html")
body.append(footer)

# 💾 حفظ الملف
with open(html_path, "w", encoding="utf-8") as f:
    f.write(str(soup))

print("✅ تمت إزالة التكرار وإضافة تذييل موحد بسطر الحقوق فقط.")
```
